```python
import geopandas as gpd
import csv
```

```python
f = open('particles.csv')
lines = f.readlines()
f.close()

# appending all lines from the file to List as shapely objects:
lines_list = []
steps = []
particles = []
for l in lines:
    
    if l[0] != '\n' and l[0] != 'X':
        # print(l.split(',')[0],l.split(',')[1].split('\n')[0])
        particles.append([l.split(',')[0], l.split(',')[1].split('\n')[0]])
        print(particles,'1')
    elif l[0] == '\n':
        print(particles,'2')
        steps.append(particles)
        particles = []
print(steps)
gs_lines = gpd.GeoSeries(lines_list)
```
